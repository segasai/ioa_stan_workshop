```python
import numpy as np 
import pystan
import pickle 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
M = pystan.StanModel('spl.stan')
```

```python
data=pickle.load(open('spl.data.pickle','rb'))
```

```python
plt.errorbar(data['x'],data['y'],data['yerr'],fmt='.')
```

```python
res = M.sampling(data=data, init='random', chains=5, iter=1000, sample_file='output.csv')
```

```python
yeval = res.extract()['yeval']
```

```python
plt.plot(data['x'], data['y'], '.')
for i in range(100):
    plt.plot(data['xeval'], yeval[i,:], color='red',alpha=0.01)
```
